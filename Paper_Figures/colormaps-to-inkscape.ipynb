```python
import colormaps as cmaps
import numpy as np
```

```python
def rgba_to_hex(rgba):
    r, g, b = [int(255 * c) for c in rgba[:3]]
    return '#{:02x}{:02x}{:02x}'.format(r, g, b)
```

```python
index = 25
cmaps.bamako.cut(0.05, "right").discrete(index)
```

```python
# Get the discrete colormap with n = index colours
colors = cmaps.bamako.cut(0.05, "right")(np.linspace(0, 1, index))

# Get the rgb colours for each and transform into hex
for i in range(index):
    rgba = colors[i]
    hex_color = rgba_to_hex(rgba)
    print(hex_color)
```
